```python
import pandas as pd
excel_file = "IMDB-Movie-Data-xls.xls"
movies = pd.read_excel(excel_file)
```

```python
movies.head(1) #Display number of rows required
```

```python
#sheetname either takes name of sheet or sheet number starting from 0
#index_col takes column nummber to set it as an index
movies_sheet1 = pd.read_excel(excel_file, sheet_name=0, index_col=2)
movies_sheet1.head()
```

```python
movies_sheet1 = pd.read_excel(excel_file, sheetname=0, index_col=0)
movies_sheet1.head()
```

```python
#reading sheet2
movies_sheet2 = pd.read_excel(excel_file, sheetname=1, index_col=0)
movies_sheet2.head()
```

```python
#reading sheet3
movies_sheet3 = pd.read_excel(excel_file, sheetname=2, index_col=0)
movies_sheet3.head()
```

```python
#concat the three dataframes created
#This way you can combine data from different sheets into one data structure
movies1 = pd.concat([movies_sheet1, movies_sheet2, movies_sheet3])
```

```python
#checking rows & columns present in our new combined data structure
movies.shape
```

```python
#The above procedure opened spreadsheet 3 times whch will become laggy for large files
# use ExcelFile class for multiple sheets of very large excel files to create a single data structure(dataframe) in the end
import sys
xlsx = pd.ExcelFile(excel_file)
movies_sheets = []
for sheet in xlsx.sheet_names:
    movies_sheets.append(xlsx.parse(sheet))
movies2 = pd.concat(movies_sheets)
```
